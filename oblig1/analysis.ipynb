```python
def viz_glcm(img, dx=1, dy=0, levels=5):
    n = slice(0, 256)
    s = slice(256, 512)
    w = n
    e = s
    fig = plt.figure()
    nw1 = glcm.patch_glcm(img[n, w], levels=levels, dx=dx, dy=dy)
    ne1 = glcm.patch_glcm(img[n, e], levels=levels, dx=dx, dy=dy)
    sw1 = glcm.patch_glcm(img[s, w], levels=levels, dx=dx, dy=dy)
    se1 = glcm.patch_glcm(img[s, e], levels=levels, dx=dx, dy=dy)
    plt.subplot(221)
    im = plt.imshow(nw1, interpolation="none")
    plt.subplot(222)
    im = plt.imshow(ne1, interpolation="none")
    plt.subplot(223)
    im = plt.imshow(sw1, interpolation="none")
    plt.subplot(224)
    im = plt.imshow(se1, interpolation="none")

    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    fig.colorbar(im, cax=cbar_ax)
```
